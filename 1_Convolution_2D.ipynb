Import libraries and image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("Input\Sheep.jpeg")
plt.imshow(img)
```

Edge detection filter

```python
kernel = np.array([[-1 , -1 , -1],
                   [-1 ,  8 , -1],
                   [-1 , -1 , -1]])
edges_img = cv2.filter2D(img, -1, kernel)
edges_img = np.hstack((img, edges_img))
plt.imshow(edges_img)
cv2.imwrite("Output/Edges_Sheep.jpeg", edges_img)
```

Sharpening filter

```python
kernel = np.array([[0  , -1 ,  0],
                   [-1 ,  5 , -1],
                   [0  , -1 ,  0]])
img_sharped = cv2.filter2D(img, -1, kernel)
img_sharped = np.hstack((img, img_sharped))
plt.imshow(img_sharped)
cv2.imwrite("Output/Sharped_Sheep.jpeg", img_sharped)
```

Emboss filter

```python
kernel = np.array([[-2 , -1 ,  0],
                   [-1 ,  1 ,  1],
                   [0  ,  1 ,  2]])
embossed_car = cv2.filter2D(img, -1, kernel)
embossed_car = np.hstack((img, embossed_car))
plt.imshow(embossed_car)
cv2.imwrite("Output/Embossed_Sheep.jpeg", embossed_car)
```

Identity filter

```python
kernel = np.array([[0  ,  0 ,  0],
                   [0  ,  1 ,  0],
                   [0  ,  0 ,  0]])
identity_car = cv2.filter2D(img, -1, kernel)
identity_car = np.hstack((img, identity_car))
plt.imshow(identity_car)
cv2.imwrite("Output/Identity_Sheep.jpeg", identity_car)
```

5*5 Edge detection filter

```python
kernel = np.array([[-1  ,  -1 ,  -1, -1, -1],
                   [-1  ,  -1 ,  4, -1, -1],
                   [-1  ,  4 ,  4, 4, -1],
                   [-1, -1, 4, -1, -1],
                   [-1, -1, -1, -1, -1]])
edge_img2 = cv2.filter2D(img, -1, kernel)
edge_img2 = np.hstack((img, edge_img2))
plt.imshow(edge_img2)
cv2.imwrite("Output/Edge_image_Sheep.jpeg", edge_img2)
```
